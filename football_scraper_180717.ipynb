```python
for a_tag in soup.find_all('a', href=True):
    print("Found the URL:", a_tag['href'])
```

```python
#pulling out players links from list
link_lst = []
for a_tag in soup.find_all('a', href=True): 
    link_lst.append(a_tag['href'])
link_lst
```

```python
#use re or startswith () to pull out player links from list
player_lst = []
for player in link_lst: 
    #if re.match('/en/players/'):
     if player.startswith('/en/players/'):
        player = 'https://fbref.com' + player
        player_lst.append(player)
```
