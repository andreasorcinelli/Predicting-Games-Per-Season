```python
#agregating data
agg_data = []

for file in os.listdir(directory):
    if file.endswith('.pkl'):
        with open(file , 'rb') as picklefile:
            ind_data = pickle.load(picklefile)
            agg_data.append(ind_data)

df = pd.concat(agg_data)
```
